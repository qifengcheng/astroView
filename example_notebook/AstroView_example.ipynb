# AstroView: A Guide to Visualizing Asteroids

Welcome to this guide for the `AstroView` package! This notebook focuses on the package's primary strength: visualizing asteroids.

We will walk through practical examples of how to:
1.  Plot the 3D orbit of a major asteroid like **4 Vesta**.
2.  Find an asteroid's precise location in the night sky from a famous observatory in the Northern Hemisphere.
3.  See how that asteroid's position changes when viewed from an observatory in the Southern Hemisphere.

Let's get started!


```python
# Setup
# First, let's import the necessary class and libraries.
# Make sure you have astroView, skyfield, astroquery, plotly, and matplotlib installed.

# unmark followings if you don't have them installed
# ! pip install AstroView
# ! pip install skyfield
# ! pip install astroquery
# ! pip install plotly
# ! pip install matplotlib

from astroView import main
from astroView.main import AsteroidVisualizer


# Create an instance of our visualizer.
# This will download the necessary ephemeris data (`de421.bsp`) on the first run.
# The data will be saved in a new `skyfield_data` directory.
print("Initializing AsteroidVisualizer...")
visualizer = AsteroidVisualizer()
print("Initialization complete.")
```


# Example 1 - Plotting a 3D Asteroid Orbit

## Example 1: Plotting the Orbit of 4 Vesta

The `plot_heliocentric_orbits_3D` method is perfect for visualizing how an asteroid orbits the Sun relative to Earth. Let's plot the path of **4 Vesta**, one of the largest objects in the asteroid belt.


```python
# Example 1
# Let's plot the orbit of Vesta for the entire year of 2025.
print("Generating 3D orbit plot for 4 Vesta...")
fig_3d = visualizer.plot_heliocentric_orbits_3D(
    object_id='Vesta', 
    start='2025-01-01', 
    stop='2025-12-31', 
    step='1d'
)

fig_3d
```

# Example 2 - Finding an Asteroid from Kitt Peak Observatory

## Example 2: Finding Asteroid 'Pallas' from a US Observatory

The `visualize_skyview` method can show you exactly where to look for an asteroid from a specific location on Earth. We'll use the Minor Planet Center (MPC) code for our observatory.

Let's try to find asteroid **Pallas** from the **Kitt Peak National Observatory** in Arizona, USA (MPC Code: `691`).


```python
# Example 2 - Code 
# Let's find Pallas in the sky on August 15, 2025.
print("\nGenerating sky view from Kitt Peak, Arizona...")
visualizer.visualize_skyview(
    objects=['Pallas', '599'], # Let's add Jupiter for reference: 599
    obs_code='691',
    obs_time_utc='2025-08-15 05:00' # 5:00 UTC is late night in Arizona
)
```

## Example 3: A Different Perspective from the Southern Hemisphere

An observer's location dramatically changes their view of the sky. Let's look for **Pallas** at the exact same time, but from the **Cerro Tololo Inter-American Observatory** in Chile (MPC Code: `807`).

Notice how the position of Pallas changes relative to the horizon and the cardinal directions.


```python
# Example 3
# Let's find Pallas again, but this time from Chile.
print("\nGenerating sky view from Cerro Tololo, Chile...")
visualizer.visualize_skyview(
    objects=['Pallas', '599'], #599 is Jupiter!
    obs_code='807',
    obs_time_utc='2025-08-15 05:00' # 5:00 UTC is late night in Chile as well
)
```

# Fun Example - A Near-Earth Object 

## Fun Example 1: Visualizing a Close Approach

Near-Earth Objects (NEOs) are asteroids or comets whose orbits bring them into Earth's neighborhood. Let's plot the orbit of one of the most famous NEOs, **99942 Apophis**, to see just how close its orbit is to our own.


```python
# Fun Example 1 
print("\nGenerating 3D orbit plot for NEO Apophis...")
fig_neo = visualizer.plot_heliocentric_orbits_3D(
    object_id='Apophis', 
    start='2025-01-01', 
    stop='2025-12-31', 
    step='1d'
)
fig_neo
```

# Fun Example - Your Local Sky

## Fun Example 2: What's in the Sky Above You?

Let's find the asteroid **Juno** from an observatory close to your location in Evanston, IL. We'll use the historic **Yerkes Observatory** in Wisconsin (MPC Code: `754`). We'll check the sky for right now!


```python
# Fun Example 2 
# Note: 2:06 PM CDT on Aug 7, 2025 is 19:06 UTC.
# The sun will be up, so the asteroid won't be visible, but we can still see its position!
print("\nGenerating local sky view from Yerkes Observatory...")
visualizer.visualize_skyview(
    objects=['-61', 'Sun'],    #Juno is -61
    obs_code='754',
    obs_time_utc='2025-08-07 19:06'
)
```
